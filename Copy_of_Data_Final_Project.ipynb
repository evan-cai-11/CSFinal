```python
import pandas as pd
import plotly.express as px

sheet_url = "https://docs.google.com/spreadsheets/d/1g-Z-BtWeH3MVYdKwSHkGiu1ndC88AWAtH5BVaYIG274/gviz/tq?tqx=out:csv"

df = pd.read_csv(sheet_url)

fig = px.scatter(df, x = "Receptions", y = "Targets", hover_data = {"Name":True, "Receptions":True, "Targets":True, "Receptions to Targets Percentage":False, "Receptions to Targets (Percentage)":True}, color = "Receptions to Targets Percentage", title = "Targets and Receptions", animation_frame = "Year", animation_group = "Receptions", range_x = [0, 190], range_y = [0, 240])

fig.show()
```
